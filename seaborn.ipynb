### annotation

```python
import matplotlib.pyplot as plt
def annotate(data, **kws):
    n = len(data)
    ax = plt.gca() ##获得当前图表
    ax.text(.1, .6, f"N = {n}", transform=ax.transAxes)
#transform=ax.transAxes 整个代码表示坐标是相对于轴边界框给出的，
#（0，0）是轴的左下角，（1，1）是右上角
g = sns.FacetGrid(tips, col="time")
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
g.map_dataframe(annotate)
```

### label/title/x&ylim/x&yticks/save figure

```python
g = sns.FacetGrid(tips, col="sex", row="time", margin_titles=True)
g.map_dataframe(sns.scatterplot, x="total_bill", y="tip")
g.set_axis_labels("Total bill ($)", "Tip ($)")
g.set_titles(col_template="{col_name} patrons", row_template="{row_name}")
g.set(xlim=(0, 60), ylim=(0, 12), xticks=[10, 30, 50], yticks=[2, 6, 10])
g.tight_layout()
g.savefig("facet_plot.png")

plt.suptitle('one for all')
```

```python
sns.countplot(x="class", hue="who", data=titanic)
```

```python
sns.countplot(y="class", hue="who", data=titanic)
```

```python
sns.countplot(x="who", data=titanic, palette="Set3")
```

```python
sns.countplot(x="who", data=titanic,
                   facecolor=(0, 0, 0, 0),
                   linewidth=5,
                   edgecolor=sns.color_palette("dark", 3))
```

```python
sns.catplot(x="class", hue="who", col="survived",
                data=titanic, kind="count",
                height=4, aspect=.7)
```

## 直接画dataframe中的obs

```python
flights = sns.load_dataset("flights")
plt.figure(figsize = (20,15))
flights = flights.pivot("month", "year", "passengers")
ax = sns.heatmap(flights)
```

```python
plt.figure(figsize = (20,15))
ax = sns.heatmap(flights, linewidths=.5)
```

```python
plt.figure(figsize = (20,15))
sns.heatmap(flights, cmap="YlGnBu")
```

```python
plt.figure(figsize = (15,10))
data = np.random.randn(50, 20)
ax = sns.heatmap(data, xticklabels=2, yticklabels=False)
```

```python
sns.heatmap(flights, cbar=False)
```

```python
#Use different axes for the colorbar:
plt.figure(figsize = (20,15))
grid_kws = {"height_ratios": (.9, .05), "hspace": .3}
f, (ax, cbar_ax) = plt.subplots(2, gridspec_kw=grid_kws)
ax = sns.heatmap(flights, ax=ax,
                 cbar_ax=cbar_ax,
                 cbar_kws={"orientation": "horizontal"})
```

```python
#Use a mask to plot only part of a matrix
corr = np.corrcoef(np.random.randn(10, 200))
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(7, 5))
    ax = sns.heatmap(corr, mask=mask, vmax=.3, square=True)
```
